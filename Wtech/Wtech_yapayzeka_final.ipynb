### Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoding_columns = ['Gender', 'CALC', 'FAVC', 'SCC', 'SMOKE', 'family_history_with_overweight', 'CAEC', 'MTRANS']

# Her bir sütunu işleyerek Label Encoding uygulama
label_encoder = LabelEncoder()
transformed_data = df.copy()

for column in encoding_columns:
    transformed_data[column] = label_encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
transformed_data.head()
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = transformed_data.iloc[:, 0:16].values
type(x)
y = transformed_data.iloc[:, 16].values
type(y)
y
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
for i, word in enumerate(le.classes_):
    print(f"{word} -> {i}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
giris_katmani = X_train.shape[1]
print(f"input_dim: {giris_katmani}")

model = Sequential()
model.add(Dense(64, input_dim=giris_katmani, activation='relu'))
model.add(Dense(32, activation='relu', name='gizli_katman1'))
model.add(Dense(16, activation='relu', name='gizli_katman2'))
model.add(Dense(7, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary() # modelin özeti

model.fit(X_train, y_train, epochs=50, batch_size=10, validation_data=(X_test, y_test))
```

## Modelin Kaydedilmesi Ve Yüklenmesi

```python
# Modelin kaydedilmesi
model.save('models/obesity.h5')
```

```python
# Modelin yüklenmesi
from tensorflow.keras.models import load_model
yuklenen_model = load_model('models/obesity.h5')

# Modelin test edilmesi
yuklenen_model.evaluate(X_test, y_test)
```
